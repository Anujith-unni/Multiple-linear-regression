```python
ipl_auction_encoded_df = pd.get_dummies( ipl_auction_df[X_features], 
columns = categorical_features, 
drop_first = True )
```

```python
ipl_auction_encoded_df.columns
```

```python
# all the above columns will be taken as independent variables i.e x features
X_features = ipl_auction_encoded_df.columns
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
X = sm.add_constant(ipl_auction_encoded_df) 
# Dependent Variable is Sold price of the player
Y = ipl_auction_df['SOLD PRICE'] 
train_X, test_X, train_y, test_y = train_test_split( X ,Y, train_size = 0.8, random_state = 42 )
```

```python
ipl_model_1 = sm.OLS(train_y, train_X).fit() 
ipl_model_1.summary2()
```

```python
# VIF stands for variation inflation factor to check mutlicolinarity between independent variables
# interview question
from statsmodels.stats.outliers_influence import variance_inflation_factor 
def get_vif_factors( X ): 
 X_matrix = X.to_numpy() 
 vif = [ variance_inflation_factor( X_matrix, i ) for i in range( X_matrix.shape[1] ) ] 
 vif_factors = pd.DataFrame() 
 vif_factors['column'] = X.columns 
 vif_factors['vif'] = vif 
 return vif_factors
```

```python
vif_factors = get_vif_factors( X[X_features] ) 
vif_factors
# if vif value is higher than 5 then x variable is a villian in our analysis because they are corelated to each other
# because of correlated values model will bring high r square which is wrong
# interview question
```

```python
columns_with_large_vif = vif_factors[vif_factors.vif > 4].column 
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure( figsize = (12,10) ) 
sn.heatmap( X[columns_with_large_vif].corr(), annot = True ); 
plt.title( "Heatmap depicting correlation between features");
```

```python
# Remove features with high coorelation means vif >4
columns_to_be_removed = ['T-RUNS', 'T-WKTS', 'RUNS-S', 'HS', 
'AVE', 'RUNS-C', 'SR-B', 'AVE-BL', 
'ECON', 'ODI-SR-B', 'ODI-RUNS-S', 'AGE_2', 'SR-BL']
```

```python
X_new_features = list( set(X_features) - set(columns_to_be_removed))
```

```python
get_vif_factors( X[X_new_features] ) 
```

```python
# if p value is less than 0.05 then there is a significant impact of feature on dependent variable
# Hypothesis Ho There is no signifiocant relationship
# if P Values is less than 0.05 we reject the statement
#Interview question

train_X = train_X[X_new_features] 
ipl_model_2 = sm.OLS(train_y, train_X).fit() 
ipl_model_2.summary2()
```
